#Plotly: Caso de uso!!

## Invocamos a las librerias necesarias: 
####zipfile y request son usadas solo para poder cargar inicialmente el dataset de ejemplo

```python
import pandas as pd
import plotly.express as px

import zipfile # Utilizado solamente para la descarga del dataset
import requests # Utilizado solamente para la descarga del dataset
```

## Cargamos datos de ejemplo: dataset de vino tinto en Portugal

```python
# Descargar el archivo ZIP desde la URL
url = 'http://www3.dsi.uminho.pt/pcortez/wine/winequality.zip'
response = requests.get(url)
zip_file = 'winequality.zip'

# Guardar el archivo ZIP en el disco
with open(zip_file, 'wb') as f:
    f.write(response.content)

# Extraer el archivo CSV del ZIP
with zipfile.ZipFile(zip_file, 'r') as z:
    csv_file = z.extract('winequality/winequality-red.csv')

# Leer el archivo CSV
data = pd.read_csv(csv_file, sep = ";")

# Visualizar los primeros registros del DataFrame
print(data.shape)
print(data.describe())
```

```python
data.columns
```

## Generamos algunos graficos simples con plotly!
#### Buscamos explorar la relacion entre algunas de las variables y la calidad del vino 

#### Dispersion

```python
fig = px.scatter(data, x='sulphates', y='quality', color='alcohol', trendline='lowess', 
                 title='Relaci√≥n entre Sulphates, Quality y alcohol',
                 labels={'sulphates': 'Sulphates', 'quality': 'Quality', 'alcohol': 'alcohol'},
                 trendline_color_override='darkblue',
                 template='plotly_white')
fig.show()
```

#### Scatter 3d!

```python
fig = px.scatter_3d(data, x='alcohol', y='sulphates', z='citric acid', color='quality')
fig.show()
```

#### Mapa de calor segun correlacion!

```python
corr_matrix = data.corr()

fig = px.imshow(corr_matrix, x=corr_matrix.columns, y=corr_matrix.columns, color_continuous_scale='Reds')
fig.show()
```
